# Відповідь на завдання 3
- застосуйте методи класу до всіх обєктів a1-a3 

```python
class MyAnimals:
    def __init__(self, animals, name, age, mass = None) -> None:
        print(f"В мене є тваринка {animals} яку звати {name} які {age} років")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass==None else mass
    
    def speak(self):
        if self.animals == "Кіт":
            return f"Моя Тваринка {self.animals} говорить Мяууу"
        elif self.animals == "Собака":
            return f"Моя Тваринка {self.animals} говорить Гавв"
        else:
            return f"Моя Тваринка {self.animals} не розмовляє"
    
    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.animals} прожила зі мною один рік"

a1 = MyAnimals("Рибки", "Неони", 1)
a2 = MyAnimals("Кіт", "Мурка", 5, 10)
a3 = MyAnimals(animals="Собака", name="Рекс", age=2)


print("\nРибки")
print(f"Голос: {a1.speak()}")     
print(f"Рік пройшов: {a1.add_one_year()}")  
print(f"Перевірка віку a1: {a1.age} років")

print("\nКіт")
print(f"Голос: {a2.speak()}")
print(f"Рік пройшов: {a2.add_one_year()}")
print(f"Перевірка віку a2: {a2.age} років")


print("\nСобака")
print(f"Голос: {a3.speak()}")
print(f"Рік пройшов: {a3.add_one_year()}")
print(f"Перевірка віку a3: {a3.age} років")
```

# Відповідь на завдання 4
- задайте декілька обєктів в циклі та виведіть глобальну змінну

```python
class MyAnimals:
    total_animals = 0
    total_animal_types = set()

    def __init__(self, animals, name, age, mass=None) -> None:
        print(f"В мене є тваринка {animals} яку звати {name} якій {age} рік/роки")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass is None else mass

        MyAnimals.total_animal_types.add(animals)
        MyAnimals.total_animals += 1

data = [
    ("Рибки", "Неони", 1),
    ("Рибки", "Гупії", 1),
    ("Рибки", "Соміки", 1),
    ("Кіт", "Мурка", 5, 10)
]

# створення обєктів у циклі
animals = []
for item in data:
    animals.append(MyAnimals(*item))

print(f"\nВ мене зараз є тваринки {list(MyAnimals.total_animal_types)}, їх кількість {MyAnimals.total_animals}")

# доступ з обєкта
a1 = animals[0]
print(f"Я доступаюсь до змінної класу з обєкта: {a1.total_animal_types}")

# перезапис змінної класу ЛОКАЛЬНО в обєкті
a1.total_animal_types = {"Собака"}

print(f"Тепер у об’єкта a1: {a1.total_animal_types}")
print(f"А у класу залишилось: {MyAnimals.total_animal_types}")
```

# Відповідь на завдання 5
- викличте статичний метод для кожного обєкта a1-a2

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
    
    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.animals} прожила за мною один рік"

    @staticmethod
    def call_pet():
        return "Тваринка біжить до нас"


a1 = MyAnimals("Рибки", "Гупії", 1)
a2 = MyAnimals("Кіт", "Мурка", 5)

# Виклик статичного методу з обєктів
print(f"a1 викликає статичний метод: |{a1.call_pet()}|")
print(f"a2 викликає статичний метод: |{a2.call_pet()}|")

# Виклик статичного методу з класу
print(f"Клас викликає статичний метод: |{MyAnimals.call_pet()}|")

# Приклад виклику метода з self
print(f"Викликаю метод з обєкту |{a1.add_one_year()}| та з класу |{MyAnimals.add_one_year(a1)}|")
```

# Відповідь на завдання 6
- додайте ще одну properties та виведіть її

```python
from datetime import datetime

class MyAnimals:
    max_age = 20

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    @property
    def live_untill(self):
        return datetime.today().year - self.age + self.max_age

    #Нова властивість
    @property
    def is_adult(self):
        return self.age >= 1

    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.animals} прожила за мною один рік"


a1 = MyAnimals("Кіт", "Мурка", 5)

print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року")

#вивід
print(f"Чи {a1.name} доросла? — {a1.is_adult}")

a1.add_one_year()

print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року")
print(f"Чи {a1.name} доросла? — {a1.is_adult}")
```

# Відповідь на завдання 7
- допишіть метод класу який буде превіряти що атрибут name заданий з великої букви 

```python
class MyAnimals:
    total_csv = int(0)
    total_objects = int(0)
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        MyAnimals.total_objects += 1

    @classmethod
    def from_csv(cls, csv_data:str):
        cls.total_csv += 1
        animal, name, age = csv_data.split(",")
        return cls(animal, name, age)

    # Метод 
    def is_name_capitalized(name:str):
        return name[0].isupper()


a1 = MyAnimals.from_csv("Кіт,Мурка,5")
a2 = MyAnimals("Собака", "Рекс", 2)

print(MyAnimals.is_name_capitalized(a1.name))  # Перевірочка
print(MyAnimals.is_name_capitalized(a2.name))  # Перевірочка

print(f"Мій {a1.animals}, звати {a1.name}, має {a1.age} років")
print(f"Мій {a2.animals}, звати {a2.name}, має {a2.age} років")
print(f"Загальна кількість обєктів {MyAnimals.total_objects} а створених з csv формату {MyAnimals.total_csv}")
```

### Магічні методи

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def __repr__(self) -> str:
        return f'MyAnimals("{self.animals}", "{self.name}", {self.age})'

    def __str__(self) -> str:
        return f"Це є обєкт {self.__repr__()}"

a1 = MyAnimals("Собака", "Рекс", 2)
print(a1)

print(repr(a1))

str(a1)
```
