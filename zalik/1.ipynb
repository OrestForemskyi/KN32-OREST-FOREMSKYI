- створити property для виведення результату
- створити метод str який буде виводити Ім'я та Групу студента що виконує завдання
- у циклі створити мінімум 2-3 об'єкти класу з різними вхідними даними

```python
class StringReverser:
    def __init__(self, text):
        self.text = text
        self._reversed = None

    def reverse(self):
        self._reversed = self.text[::-1]

    @property
    def reversed_string(self):
        if self._reversed is None:
            self.reverse()
        return self._reversed

    def __str__(self):
        return "Студент: Орест Форемський, група КНк-32"
    
strings = [
    "Це тестова стрічка",
    "Обєктно орієнтоване програмування",
    "Python"
]

objects = []

for s in strings:
    obj = StringReverser(s)
    objects.append(obj)

for obj in objects:
    print(obj)
    print("Оригінальний рядок:", obj.text)
    print("Зворотний рядок:", obj.reversed_string)
    print("-" * 40)


```
