### Функціональний підхід

– створюємо картку просто як словник    
– робимо функцію, що змінює її стан   
– все на dict і функціях   

```python
def create_card(name, damage, durability):
    return {
        "name": name,
        "damage": damage,
        "durability": durability
    }

def hit(card):
    card["durability"] -= 1

sword = create_card("Sword", 10, 5)
hit(sword)
sword
```

### Динамічні об’єкти Python
Можна створити пустий об’єкт і додати йому атрибути так само, як у словник.

```python
class Empty:
    pass

item = Empty()
item.name = "Sword"
item.damage = 15
item.durability = 4

item.durability -= 1
item.durability
```

### Перший приклад з init

```python
class Weapon:
    def __init__(self):
        self.name = "Sword"
        self.damage = 10
        self.durability = 5

sword = Weapon()
sword.durability -= 1
sword.durability
```

### Використання параметрів у конструкторі

```python
class Weapon:
    def __init__(self, name, damage, durability):
        self.name = name
        self.damage = damage
        self.durability = durability

sword = Weapon("Iron Sword", 12, 8)
sword.damage
```

### @property можна зробити віртуальний атрибут, який обчислюється:
як звичайна змінна, але всередині функція.

```python
class Weapon:
    def __init__(self, damage, speed):
        self.damage = damage
        self.speed = speed

    @property
    def dps(self):
        return self.damage * self.speed

axe = Weapon(10, 1.2)
axe.dps
```

### Інкапсуляція (ховаємо внутрішні поля)
чому краще клас ніж словник

```python
class Account:
    def __init__(self, money):
        self.__money = money

    def deposit(self, amount):
        self.__money += amount

    def get_money(self):
        return self.__money

acc = Account(100)
acc.deposit(50)
acc.get_money()
```

### Методи, які змінюють об’єкт (hit / attack)

```python
class Weapon:
    def __init__(self, name, dmg, durability):
        self.name = name
        self.dmg = dmg
        self.durability = durability

    def hit(self, target):
        self.durability -= 1
        target.hp -= self.dmg

class Enemy:
    def __init__(self, hp):
        self.hp = hp

orc = Enemy(30)
sword = Weapon("Sword", 8, 5)

sword.hit(orc)
orc.hp, sword.durability
```

### Наслідування (Inheritance)

```python
class Animal:
    def speak(self):
        return "Якась тварина видає звук"

class Cat(Animal):
    def speak(self):
        return "Кіт каже: Мяу"

class Dog(Animal):
    def speak(self):
        return "Пес каже: Гав"
```
