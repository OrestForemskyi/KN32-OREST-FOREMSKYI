## 1. –ó–Ω–∞–π–æ–º—Å—Ç–≤–æ –∑ –û—Å–Ω–æ–≤–Ω–∏–º–∏ –¢–∏–ø–∞–º–∏ –î–∞–Ω–∏—Ö
Python —î –º–æ–≤–æ—é –∑ –¥–∏–Ω–∞–º—ñ—á–Ω–æ—é —Ç–∏–ø—ñ–∑–∞—Ü—ñ—î—é. –£—Å—ñ –∑–º—ñ–Ω–Ω—ñ —î –æ–±'—î–∫—Ç–∞–º–∏, —â–æ –¥–æ–∑–≤–æ–ª—è—î –∑–∞—Å—Ç–æ—Å–æ–≤—É–≤–∞—Ç–∏ –¥–æ –Ω–∏—Ö –≤–±—É–¥–æ–≤–∞–Ω—ñ –º–µ—Ç–æ–¥–∏. –ú–∏ –ø—Ä–∞–∫—Ç–∏–∫—É—î–º–æ—Å—è –∑ –æ—Å–Ω–æ–≤–Ω–∏–º–∏ —Ç–∏–ø–∞–º–∏ (str, int, float) —Ç–∞ –∫–æ–ª–µ–∫—Ü—ñ—è–º–∏ (list, tuple, set, dict).

```
str (–°—Ç—Ä—ñ—á–∫–∞): –ù–µ–∑–º—ñ–Ω–Ω–∞ –ø–æ—Å–ª—ñ–¥–æ–≤–Ω—ñ—Å—Ç—å —Å–∏–º–≤–æ–ª—ñ–≤, –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –¥–ª—è —Ç–µ–∫—Å—Ç—É.

int / float: –ß–∏—Å–ª–æ–≤—ñ —Ç–∏–ø–∏ –¥–ª—è —Ü—ñ–ª–∏—Ö —á–∏—Å–µ–ª —Ç–∞ —á–∏—Å–µ–ª –∑ –ø–ª–∞–≤–∞—é—á–æ—é –∫—Ä–∞–ø–∫–æ—é.

list (–°–ø–∏—Å–æ–∫): –í–ø–æ—Ä—è–¥–∫–æ–≤–∞–Ω–∞, –∑–º—ñ–Ω–Ω–∞ –∫–æ–ª–µ–∫—Ü—ñ—è.

tuple (–ö–æ—Ä—Ç–µ–∂): –í–ø–æ—Ä—è–¥–∫–æ–≤–∞–Ω–∞, –Ω–µ–∑–º—ñ–Ω–Ω–∞ –∫–æ–ª–µ–∫—Ü—ñ—è.

set (–ù–∞–±—ñ—Ä): –ù–µ–≤–ø–æ—Ä—è–¥–∫–æ–≤–∞–Ω–∞ –∫–æ–ª–µ–∫—Ü—ñ—è —É–Ω—ñ–∫–∞–ª—å–Ω–∏—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤.

dict (–°–ª–æ–≤–Ω–∏–∫): –ó–±–µ—Ä—ñ–≥–∞—î –¥–∞–Ω—ñ —è–∫ –ø–∞—Ä–∏ –∫–ª—é—á: –∑–Ω–∞—á–µ–Ω–Ω—è (key: value).
```

````python
text_var = "–¶–µ —Ä—è–¥–æ–∫ –¥–ª—è —Ç–µ—Å—Ç—É–≤–∞–Ω–Ω—è"
int_var = 42
float_var = 3.1415

# –ö–æ–ª–µ–∫—Ü—ñ—ó
list_var = [1, "a", 10.5, text_var]
dict_var = {"id": 1, "–Ω–∞–∑–≤–∞": "–ï–ª–µ–º–µ–Ω—Ç", text_var: int_var}
tuple_var = ("–∫–æ—Ä—Ç–µ–∂", int_var)
set_var = {1, 2, 2, "a", "b"} 

print(f"–¢–µ–∫—Å—Ç (str): {text_var}, –î–æ–≤–∂–∏–Ω–∞: {len(text_var)}")
print(f"–¶—ñ–ª–µ —á–∏—Å–ª–æ (int): {int_var}, –¢–∏–ø: {type(int_var)}")
print(f"–°–ø–∏—Å–æ–∫ (list): {list_var}")
print(f"–ù–∞–±—ñ—Ä (set) –ø—ñ—Å–ª—è –≤–∏–¥–∞–ª–µ–Ω–Ω—è –¥—É–±–ª—ñ–∫–∞—Ç—ñ–≤: {set_var}")

# –ü—Ä–∞–∫—Ç–∏—á–Ω–µ –∑–∞–≤–¥–∞–Ω–Ω—è: –°–ª–∞–π—Å–∏–Ω–≥ —Ç–∞ –º–µ—Ç–æ–¥–∏
print(f"\n–°–ª–∞–π—Å–∏–Ω–≥ list_var[1:3]: {list_var[1:3]}")
print(f"–í–µ—Ä—Ö–Ω—ñ–π —Ä–µ–≥—ñ—Å—Ç—Ä —Ç–µ–∫—Å—Ç—É: {text_var.upper()}")
````

## 2/3 –í–±—É–¥–æ–≤–∞–Ω—ñ –∑–º—ñ–Ω–Ω—ñ, —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ –∑–∞—Ä–µ–∑–µ—Ä–≤–æ–≤–∞–Ω—ñ —Å–ª–æ–≤–∞
- –≤–±—É–¥–æ–≤–∞–Ω—ñ –∑–º—ñ–Ω–Ω—ñ –Ω–µ–º–æ–∂–Ω–∞ –ø–µ—Ä–µ–ø—Ä–∏—Å–≤–æ—é–≤–∞—Ç–∏ - —Ü–µ –±—É–¥–µ –ø–æ–º–∏–ª–∫–∞ `SyntaxError: cannot assign to True`
- None - –æ—Å–æ–±–ª–∏–≤–∞ –∑–º—ñ–Ω–Ω—ñ —è–∫–∞ –æ–∑–Ω–∞—á–∞—î –Ω—ñ—â–æ —ñ –Ω–µ—é –∑–∞–∑–≤–∏—á–∞–π —ñ–Ω—ñ—Ü—ñ–∞–ª—ñ–∑—É—é—Ç—å –∑–º—ñ–Ω–Ω—ñ

–í–±—É–¥–æ–≤–∞–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—ó: –§—É–Ω–∫—Ü—ñ—ó, –¥–æ—Å—Ç—É–ø–Ω—ñ –≥–ª–æ–±–∞–ª—å–Ω–æ, –Ω–∞–ø—Ä–∏–∫–ª–∞–¥, abs() (–∞–±—Å–æ–ª—é—Ç–Ω–µ –∑–Ω–∞—á–µ–Ω–Ω—è), len() (–¥–æ–≤–∂–∏–Ω–∞), sorted() (—Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è).

````python
import keyword
    
# –í–±—É–¥–æ–≤–∞–Ω—ñ –∫–æ–Ω—Å—Ç–∞–Ω—Ç–∏
print(f"1. –ë—É–ª–µ–≤–µ –∑–Ω–∞—á–µ–Ω–Ω—è True: {True}")
print(f"2. –ö–æ–Ω—Å—Ç–∞–Ω—Ç–∞ –≤—ñ–¥—Å—É—Ç–Ω–æ—Å—Ç—ñ –∑–Ω–∞—á–µ–Ω–Ω—è None: {None}")
print(f"3. –ü–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è –∑–Ω–∞—á–µ–Ω—å (True == 1): {True == 1}")

# –í–±—É–¥–æ–≤–∞–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—ó
print(f"\n1. abs(): –ê–±—Å–æ–ª—é—Ç–Ω–µ –∑–Ω–∞—á–µ–Ω–Ω—è abs(-5.7) = {abs(-5.7)}")
print(f"2. pow(): –ü—ñ–¥–Ω–µ—Å–µ–Ω–Ω—è –≤ —Å—Ç–µ–ø—ñ–Ω—å pow(3, 4) = {pow(3, 4)}")
print(f"3. sorted(): –°–æ—Ä—Ç—É–≤–∞–Ω–Ω—è [3, 1, 2] = {sorted([3, 1, 2])}")

# –ó–∞—Ä–µ–∑–µ—Ä–≤–æ–≤–∞–Ω—ñ —Å–ª–æ–≤–∞
print("\n--- –ó–∞—Ä–µ–∑–µ—Ä–≤–æ–≤–∞–Ω—ñ —Å–ª–æ–≤–∞ Python ---")
print(keyword.kwlist)
````

## 4. –¶–∏–∫–ª–∏ (`for`, `while`) —Ç–∞ –Ü—Ç–µ—Ä–∞—Ü—ñ—ó
–ü—Ä–∞–∫—Ç–∏–∫–∞ —Ü–∏–∫–ª—ñ–≤ –¥–ª—è —ñ—Ç–µ—Ä–∞—Ü—ñ—ó –ø–æ –∫–æ–ª–µ–∫—Ü—ñ—è—Ö —Ç–∞ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è `else` –ø—ñ—Å–ª—è —Ü–∏–∫–ª—É `for` (–≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è, —è–∫—â–æ —Ü–∏–∫–ª –Ω–µ –±—É–≤ –ø–µ—Ä–µ—Ä–≤–∞–Ω–∏–π `break`).

````python
items = ["–º–æ–Ω—ñ—Ç–æ—Ä", "–∫–ª–∞–≤—ñ–∞—Ç—É—Ä–∞", "–º–∏—à–∞"]
    
# –¶–∏–∫–ª FOR –∑ else
print("--- –¶–∏–∫–ª FOR –∑ else ---")
for i in range(len(items)):
    print(f"–ù–∞ –ø–æ–∑–∏—Ü—ñ—ó {i} –∑–Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è: {items[i]}")
else:
    print("–ü–µ—Ä–µ–ª—ñ–∫ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –∑–∞–≤–µ—Ä—à–µ–Ω–æ.")

# –¶–∏–∫–ª WHILE
print("\n--- –¶–∏–∫–ª WHILE –∑ —É–º–æ–≤–æ—é ---")
counter = 4
while counter > 0:
    print(f"–ó–≤–æ—Ä–æ—Ç–Ω–∏–π –≤—ñ–¥–ª—ñ–∫: {counter}")
    counter -= 1
    if counter == 2:
        print("–ü—Ä–æ–ø—É—Å–∫–∞—î–º–æ –∫—Ä–æ–∫ 2.")
        continue
````

## 5. –†–æ–∑–≥–∞–ª—É–∂–µ–Ω–Ω—è (`if/elif/else`)
–í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è —É–º–æ–≤–Ω–∏—Ö –∫–æ–Ω—Å—Ç—Ä—É–∫—Ü—ñ–π –¥–ª—è –∫–µ—Ä—É–≤–∞–Ω–Ω—è –ª–æ–≥—ñ–∫–æ—é –ø—Ä–æ–≥—Ä–∞–º–∏. –î–µ–º–æ–Ω—Å—Ç—Ä—É—î–º–æ –ø–æ–≤–Ω—É –∫–æ–Ω—Å—Ç—Ä—É–∫—Ü—ñ—é —Ç–∞ —Å—Ç–∏—Å–ª–∏–π —Ç–µ—Ä–Ω–∞—Ä–Ω–∏–π –æ–ø–µ—Ä–∞—Ç–æ—Ä.

[true_action] if [condition] else [false_action]

````python
from random import randint
score = randint(50, 100)

print(f"–ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ –æ—Ü—ñ–Ω–∫–∏: {score}")

if score >= 90:
    grade = "–í—ñ–¥–º—ñ–Ω–Ω–æ"
elif score >= 70:
    grade = "–î–æ–±—Ä–µ"
else:
    grade = "–ó–∞–¥–æ–≤—ñ–ª—å–Ω–æ"

print(f"–†–µ–π—Ç–∏–Ω–≥: {grade}")

# –¢–µ—Ä–Ω–∞—Ä–Ω–∏–π –æ–ø–µ—Ä–∞—Ç–æ—Ä
status_msg = "–£—Å–ø—ñ—à–Ω–æ" if score > 60 else "–ù–µ–æ–±—Ö—ñ–¥–Ω–æ –ø–µ—Ä–µ–∑–¥–∞—Ç–∏"
print(f"\n–°—Ç–∞—Ç—É—Å —Å—Ç—É–¥–µ–Ω—Ç–∞: {status_msg}")
````

## 6. –û–±—Ä–æ–±–∫–∞ –ü–æ–º–∏–ª–æ–∫ (`try`->`except`->`finally`)
–ö–æ–Ω—Å—Ç—Ä—É–∫—Ü—ñ—è –¥–ª—è –±–µ–∑–ø–µ—á–Ω–æ–≥–æ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –∫–æ–¥—É, –¥–µ –º–æ–∂—É—Ç—å –≤–∏–Ω–∏–∫–Ω—É—Ç–∏ –≤–∏–Ω—è—Ç–∫–∏ (–ø–æ–º–∏–ª–∫–∏). –ë–ª–æ–∫ `except` –ø–µ—Ä–µ—Ö–æ–ø–ª—é—î –ø–æ–º–∏–ª–∫—É, –∞ `finally` –≥–∞—Ä–∞–Ω—Ç—É—î –≤–∏–∫–æ–Ω–∞–Ω–Ω—è —Ñ—ñ–Ω–∞–ª—å–Ω–∏—Ö –¥—ñ–π.

try: –ë–ª–æ–∫, —è–∫–∏–π –º—ñ—Å—Ç–∏—Ç—å –ø–æ—Ç–µ–Ω—Ü—ñ–π–Ω–æ –Ω–µ–±–µ–∑–ø–µ—á–Ω–∏–π –∫–æ–¥.

except: –ü–µ—Ä–µ—Ö–æ–ø–ª—é—î —Ç–∞ –æ–±—Ä–æ–±–ª—è—î –∫–æ–Ω–∫—Ä–µ—Ç–Ω–∏–π —Ç–∏–ø –≤–∏–Ω—è—Ç–∫—É (–ø–æ–º–∏–ª–∫–∏).

finally: –í–∏–∫–æ–Ω—É—î—Ç—å—Å—è –∑–∞–≤–∂–¥–∏, –Ω–µ–∑–∞–ª–µ–∂–Ω–æ –≤—ñ–¥ —Ç–æ–≥–æ, —á–∏ –±—É–ª–∞ –ø–æ–º–∏–ª–∫–∞.

````python
divider = 0

try:
    print("–°–ø—Ä–æ–±–∞ –¥—ñ–ª–µ–Ω–Ω—è –Ω–∞ –Ω—É–ª—å...")
    result = 100 / divider
    print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {result}")
except ZeroDivisionError as e:
    print(f"–ù–µ–≤–∂–µ —Ü–µ –ø–æ–º–∏–ª–∫–∞ –¥—ñ–ª–µ–Ω–Ω—è > {e}")
    result = 0
except Exception as e:
    print(f"–í–∏–Ω–∏–∫–ª–∞ —ñ–Ω—à–∞ –ø–æ–º–∏–ª–∫–∞ > {e}")
finally:
    print(f"\n–§—ñ–Ω–∞–ª—å–Ω–∏–π –±–ª–æ–∫: –û–ø–µ—Ä–∞—Ü—ñ—è –∑–∞–≤–µ—Ä—à–µ–Ω–∞ –∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–º {result}.")
````

## 7. –ö–æ–Ω—Ç–µ–∫—Å—Ç-–º–µ–Ω–µ–¥–∂–µ—Ä (`with`)
–ö–æ–Ω—Ç–µ–∫—Å—Ç-–º–µ–Ω–µ–¥–∂–µ—Ä –∑–∞–±–µ–∑–ø–µ—á—É—î –∫–æ—Ä–µ–∫—Ç–Ω–µ –∫–µ—Ä—É–≤–∞–Ω–Ω—è —Ä–µ—Å—É—Ä—Å–∞–º–∏ (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, —Ñ–∞–π–ª–∞–º–∏), –∞–≤—Ç–æ–º–∞—Ç–∏—á–Ω–æ –≤–∏–∫–ª–∏–∫–∞—é—á–∏ –æ—á–∏—â–µ–Ω–Ω—è (`__exit__`), –Ω–∞–≤—ñ—Ç—å —è–∫—â–æ –≤—Å–µ—Ä–µ–¥–∏–Ω—ñ –±–ª–æ–∫—É –≤–∏–Ω–∏–∫–ª–∞ –ø–æ–º–∏–ª–∫–∞.

````python
file_path = "temp_report.txt"

# –ó–∞–ø–∏—Å —É —Ñ–∞–π–ª
with open(file_path, "w", encoding="utf-8") as f:
    f.write("–ó–∞–ø–∏—Å 1: –î–∞–Ω—ñ –∑ –ª–∞–±–æ—Ä–∞—Ç–æ—Ä–Ω–æ—ó.\n")
    f.write("–ó–∞–ø–∏—Å 2: –¶–µ —É with-–±–ª–æ—Ü—ñ.\n")

# –ß–∏—Ç–∞–Ω–Ω—è —Ñ–∞–π–ª—É
with open(file_path, "r", encoding="utf-8") as f:
    print("\n--- –í–º—ñ—Å—Ç —Ñ–∞–π–ª—É ---")
    for index, line in enumerate(f):
        print(f"–†—è–¥–æ–∫ {index}): {line.strip()}")
````

## 8. Python Lambdas
–õ—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—ó ‚Äî —Ü–µ –∞–Ω–æ–Ω—ñ–º–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—ó, –≤–∏–∑–Ω–∞—á–µ–Ω—ñ –≤ –æ–¥–Ω–æ–º—É —Ä—è–¥–∫—É. –í–æ–Ω–∏ –º—ñ—Å—Ç—è—Ç—å –æ–¥–∏–Ω –≤–∏—Ä–∞–∑, —Ä–µ–∑—É–ª—å—Ç–∞—Ç —è–∫–æ–≥–æ –ø–æ–≤–µ—Ä—Ç–∞—é—Ç—å. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—Ç—å—Å—è —á–∞—Å—Ç–æ –¥–ª—è —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è –∞–±–æ —Ñ—ñ–ª—å—Ç—Ä–∞—Ü—ñ—ó –¥–∞–Ω–∏—Ö, –¥–µ –Ω–µ –ø–æ—Ç—Ä—ñ–±–Ω–∞ –ø–æ–≤–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è `def`.

````python
def user_data_tuple(name, score):
    return name, score

# –õ—è–º–±–¥–∞ –¥–ª—è —Ñ–æ—Ä–º–∞—Ç—É–≤–∞–Ω–Ω—è –≤–∏–≤–æ–¥—É
formatter = lambda item, value: f'–ü–∞—Ä–∞–º–µ—Ç—Ä: {item}, –ó–Ω–∞—á–µ–Ω–Ω—è: {value}'

print(f"–õ—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—è: {formatter('–ö—É—Ä—Å', 3)}")

# –í–∏–∫–ª–∏–∫ –ª—è–º–±–¥–∏ –∑ —Ä–æ–∑–ø–∞–∫—É–≤–∞–Ω–Ω—è–º –∫–æ—Ä—Ç–µ–∂—É
print(f"–í–∏–∫–ª–∏–∫ –∑ —Ä–æ–∑–ø–∞–∫—É–≤–∞–Ω–Ω—è–º: {formatter(*user_data_tuple('–û–ª–µ–≥', 95))}")

# –õ—è–º–±–¥–∞ —è–∫ –∫–ª—é—á —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è
scores = [('A', 88), ('B', 92), ('C', 75)]
sorted_scores = sorted(scores, key=lambda item: item[1])
print(f"\n–í—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω–æ –∑–∞ –æ—Ü—ñ–Ω–∫–æ—é: {sorted_scores}")
````

 # –í—ñ–¥–ø–æ–≤—ñ–¥—å –≤—ñ–¥ AI –∑ –ø—Ä–∞–∫—Ç–∏—á–Ω–∏–º–∏ –∑–∞–¥–∞—á–∞–º–∏ 

### –§–æ—Ä–º–∞—Ç—É–≤–∞–Ω–Ω—è –Ü–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—ó –ø—Ä–æ –°—Ç—É–¥–µ–Ω—Ç–∞
–ó–∞–≤–¥–∞–Ω–Ω—è: –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ f-—Å—Ç—Ä—ñ—á–∫–∏ —Ç–∞ —Ä—ñ–∑–Ω—ñ —Ç–∏–ø–∏ –¥–∞–Ω–∏—Ö, —Å—Ç–≤–æ—Ä–∏—Ç–∏ —Ä—è–¥–æ–∫, —â–æ –º—ñ—Å—Ç–∏—Ç—å —ñ–º'—è, –∫—É—Ä—Å —Ç–∞ —Å–µ—Ä–µ–¥–Ω—ñ–π –±–∞–ª —Å—Ç—É–¥–µ–Ω—Ç–∞.

````python
student_name = "–û–ª–µ–∫—Å–∞–Ω–¥—Ä"
course = 3
average_score = 88.5
is_active = True

info_string = f"–°—Ç—É–¥–µ–Ω—Ç: {student_name}, –ö—É—Ä—Å: {course}, –°–µ—Ä–µ–¥–Ω—ñ–π –±–∞–ª: {average_score:.1f}, –ê–∫—Ç–∏–≤–Ω–∏–π: {is_active}"

print(info_string)
````

### –û–±—Ä–æ–±–∫–∞ –°–ø–∏—Å–∫—É –û—Ü—ñ–Ω–æ–∫ üìù
–ó–∞–≤–¥–∞–Ω–Ω—è: –°—Ç–≤–æ—Ä–∏—Ç–∏ —Å–ø–∏—Å–æ–∫ –æ—Ü—ñ–Ω–æ–∫, –∑–Ω–∞–π—Ç–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É —Ç–∞ –º—ñ–Ω—ñ–º–∞–ª—å–Ω—É, —ñ –æ–±—á–∏—Å–ª–∏—Ç–∏ —Å–µ—Ä–µ–¥–Ω—ñ–π –±–∞–ª.

````python
grades = [75, 88, 92, 60, 95, 80]

max_grade = max(grades)
min_grade = min(grades)
average_grade = sum(grades) / len(grades)

print(f"–°–ø–∏—Å–æ–∫ –æ—Ü—ñ–Ω–æ–∫: {grades}")
print(f"–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–∞ –æ—Ü—ñ–Ω–∫–∞: {max_grade}")
print(f"–ú—ñ–Ω—ñ–º–∞–ª—å–Ω–∞ –æ—Ü—ñ–Ω–∫–∞: {min_grade}")
print(f"–°–µ—Ä–µ–¥–Ω—ñ–π –±–∞–ª: {average_grade:.2f}")
````

### –§—ñ–ª—å—Ç—Ä–∞—Ü—ñ—è –°–ø–∏—Å–∫—É –∑–∞ –£–º–æ–≤–æ—é üîç
–ó–∞–≤–¥–∞–Ω–Ω—è: –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ —Ü–∏–∫–ª for —Ç–∞ —É–º–æ–≤—É if, —Å—Ç–≤–æ—Ä–∏—Ç–∏ –Ω–æ–≤–∏–π —Å–ø–∏—Å–æ–∫, —â–æ –º—ñ—Å—Ç–∏—Ç—å —Ç—ñ–ª—å–∫–∏ –ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞ –∑ –≤–∏—Ö—ñ–¥–Ω–æ–≥–æ —Å–ø–∏—Å–∫—É.

````python
numbers = [1, 5, 12, 7, 8, 20, 3, 10]
even_numbers = []

for num in numbers:
    if num % 2 == 0:
        even_numbers.append(num)

print(f"–í–∏—Ö—ñ–¥–Ω–∏–π —Å–ø–∏—Å–æ–∫: {numbers}")
print(f"–¢—ñ–ª—å–∫–∏ –ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞: {even_numbers}")
````

### –ü—Ä–æ—Å—Ç–∞ –§—É–Ω–∫—Ü—ñ—è: –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ –î–æ–≤–∂–∏–Ω–∏ –†—è–¥–∫–∞ ‚úÖ
–ó–∞–≤–¥–∞–Ω–Ω—è: –ù–∞–ø–∏—Å–∞—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—é, —è–∫–∞ –ø—Ä–∏–π–º–∞—î —Ä—è–¥–æ–∫ —ñ –ø–æ–≤–µ—Ä—Ç–∞—î True, —è–∫—â–æ –π–æ–≥–æ –¥–æ–≤–∂–∏–Ω–∞ –±—ñ–ª—å—à–∞ –∑–∞ 10 —Å–∏–º–≤–æ–ª—ñ–≤, —ñ False –≤ —ñ–Ω—à–æ–º—É –≤–∏–ø–∞–¥–∫—É.

````python
def is_long_string(text):
  """–ü–µ—Ä–µ–≤—ñ—Ä—è—î, —á–∏ –¥–æ–≤–∂–∏–Ω–∞ —Ä—è–¥–∫–∞ –±—ñ–ª—å—à–∞ –∑–∞ 10."""
  return len(text) > 10

test_string1 = "–ö–æ—Ä–æ—Ç–∫–∏–π —Ç–µ–∫—Å—Ç"
test_string2 = "–î—É–∂–µ –¥–æ–≤–≥–∏–π —Ç–µ–∫—Å—Ç –¥–ª—è –ø–µ—Ä–µ–≤—ñ—Ä–∫–∏"

print(f"'{test_string1}' - –¥–æ–≤–≥–∏–π? {is_long_string(test_string1)}")
print(f"'{test_string2}' - –¥–æ–≤–≥–∏–π? {is_long_string(test_string2)}")
````

### –ë–µ–∑–ø–µ—á–Ω–µ –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è –¢–∏–ø—É (Error Handling) üõ°Ô∏è
–ó–∞–≤–¥–∞–Ω–Ω—è: –°–ø—Ä–æ–±—É–≤–∞—Ç–∏ –ø–µ—Ä–µ—Ç–≤–æ—Ä–∏—Ç–∏ –µ–ª–µ–º–µ–Ω—Ç–∏ —Å–ø–∏—Å–∫—É –Ω–∞ —Ü—ñ–ª—ñ —á–∏—Å–ª–∞, –æ–±—Ä–æ–±–ª—è—é—á–∏ –º–æ–∂–ª–∏–≤—É –ø–æ–º–∏–ª–∫—É ValueError.

````python
data_to_convert = ["10", "25", "abc", "40", "xyz"]
converted_numbers = []

for item in data_to_convert:
    try:
        number = int(item)
        converted_numbers.append(number)
    except ValueError:
        print(f"–ü–æ–º–∏–ª–∫–∞: –ù–µ–º–æ–∂–ª–∏–≤–æ –ø–µ—Ä–µ—Ç–≤–æ—Ä–∏—Ç–∏ '{item}' –Ω–∞ —á–∏—Å–ª–æ. –ü—Ä–æ–ø—É—Å–∫–∞—î–º–æ.")
    finally:
        print(f"–û–±—Ä–æ–±–∫–∞ '{item}' –∑–∞–≤–µ—Ä—à–µ–Ω–∞.")


print(f"\n–ö–æ–Ω–≤–µ—Ä—Ç–æ–≤–∞–Ω—ñ —á–∏—Å–ª–∞: {converted_numbers}")
````
