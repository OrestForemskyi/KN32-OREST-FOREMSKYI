## 1. Інкапсуляція
**Інкапсуляція** - це приховування внутрішнього стану об'єкта та захист його від прямого втручання. 
- В Python для цього використовуються подвійні підкреслення `__` перед назвою змінної або методу.
- Приклад: програма, яка визначає, чи занудьгував студент на лекції (внутрішня логіка розрахунку прихована).

```python
import random

class CheckStudentBored:
    def __init__(self, name, words_threshold=50):
        self.__name = name          # Приватний атрибут
        self.__words_threshold = words_threshold  # Приватний атрибут
    
    def __is_bored(self):           # Приватний метод
        lecture_words = random.randint(0, 100)
        return lecture_words > self.__words_threshold
    
    def check(self):                # Публічний метод для взаємодії
        if self.__is_bored():
            return f"Студент {self.__name} занудьгував."
        return f"Студент {self.__name} уважно слухає."

student = CheckStudentBored("Віктор", 40)
print(student.check())
```

## 2. Наслідування
**Наслідування** - можливість створювати нові класи на основі існуючих (базових), переймаючи їхні атрибути та методи.
- Це дозволяє уникати дублювання коду.
- Приклад: базова станція живлення та її розширення до сонячної панелі.

```python
class EnergyDevice:
    def __init__(self, name, capacity):
        self.name = name
        self.capacity = capacity
        
    def info(self):
        return f"Пристрій: {self.name}, Ємність: {self.capacity}"

class SolarPanelDevice(EnergyDevice): # Наслідуємось від EnergyDevice
    def __init__(self, name, capacity, panel_area):
        super().__init__(name, capacity) # Виклик конструктора базового класу
        self.panel_area = panel_area

    def info(self):
        base_info = super().info() # Перевизначення методу з викликом базового
        return f"{base_info}, Площа панелі: {self.panel_area} м²"

solar = SolarPanelDevice("Сонячна Панель", 100, 20)
print(solar.info())
```

## 3. Поліморфізм (Duck Typing)
**Поліморфізм** - здатність об'єктів різних класів реагувати на однакові виклики методів по-різному.
- *Duck Typing*: Якщо об'єкт ходить як качка і крякає як качка, ми можемо вважати його качкою.
- Приклад: Павербанк може заряджатися від будь-якого пристрою, який має метод `charge_device`.

```python
class OutletDevice:
    def charge_device(self):
        return "Заряджаємо від розетки (швидко)"

class USBDevice:
    def charge_device(self):
        return "Заряджаємо від USB (повільніше)"

class PowerBank:
    def charge_from(self, device):
        # Павербанку байдуже, який це клас, головне, щоб був метод charge_device
        print(device.charge_device())

pb = PowerBank()
outlet = OutletDevice()
usb = USBDevice()

pb.charge_from(outlet)
pb.charge_from(usb)
```

## 4. Абстракція
**Абстракція** - виділення головних характеристик об'єкта і відкидання неважливих деталей. Задає "контракт" (інтерфейс), який мають виконувати дочірні класи.
- Використовуємо модуль `abc` (Abstract Base Classes).
- Неможливо створити екземпляр абстрактного класу.

```python
from abc import ABC, abstractmethod

# Створюємо абстрактний базовий клас для всіх LLM моделей
class LLMModel(ABC):
    @abstractmethod
    def generate_response(self, prompt):
        pass

# Конкретна реалізація
class ITCollegeLLM(LLMModel):
    def generate_response(self, prompt):
        return f"[IT College AI] Відповідь на запит: '{prompt}'"

model = ITCollegeLLM()
print(model.generate_response("Що таке ООП?"))
```
